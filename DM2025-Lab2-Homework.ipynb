## 1. Preprocessing Steps

```python
import json
from pathlib import Path
import pandas as pd
import re, html

CAND_DIRS = [Path("."), Path("./data")]
def find(name: str) -> Path:
    for d in CAND_DIRS:
        p = d / name
        if p.exists():
            return p
    raise FileNotFoundError(f"{name} not found in {', '.join(str(x) for x in CAND_DIRS)}")

#data give from kaggle
posts_path   = find("final_posts.json")
emotion_path = find("emotion.csv")
id_path      = find("data_identification.csv")
print("Using files:\n ", posts_path, "\n ", emotion_path, "\n ", id_path)

#loading posts.json
with open(posts_path, "r", encoding="utf-8") as f:
    raw = json.load(f)

def extract_rows(obj):
    rows = []
    if isinstance(obj, dict):
        for k, v in obj.items():
            rows.append({"id": str(k), "text": str(v)})
        return rows
    if isinstance(obj, list):
        for it in obj:
            if isinstance(it, dict):
                pid = it.get("id", it.get("post_id"))
                txt = it.get("text", it.get("content", it.get("body")))
                if pid is not None and txt is not None:
                    rows.append({"id": str(pid), "text": str(txt)})
                    continue
                #try nested reddit-like structure
                root = it.get("root", {})
                post = root.get("_source", {}).get("post", {})
                pid = post.get("post_id"); txt = post.get("text")
                if pid is not None and txt is not None:
                    rows.append({"id": str(pid), "text": str(txt)})
    return rows

posts_rows = extract_rows(raw if not (isinstance(raw, dict) and "posts" in raw) else raw["posts"])
assert posts_rows, "final_posts.json parsed 0 rows."
posts_df = pd.DataFrame(posts_rows, columns=["id","text"]).dropna().reset_index(drop=True)
posts_df["id"] = posts_df["id"].astype(str)

#loading emotion.csv
emo = pd.read_csv(emotion_path)
id_col  = next((c for c in emo.columns if str(c).lower() in {"id","post_id"}), None)
lab_col = next((c for c in emo.columns if str(c).lower() in {"emotion","label","target","class"}), None)
assert id_col and lab_col, f"emotion.csv must have id + emotion. Got: {list(emo.columns)}"
emotion_df = emo.rename(columns={id_col:"id", lab_col:"emotion"})[["id","emotion"]]
emotion_df["id"] = emotion_df["id"].astype(str)

#loading data_identification.csv
ids = pd.read_csv(id_path)
idc    = next((c for c in ids.columns if str(c).lower() in {"id","post_id"}), None)
splitc = next((c for c in ids.columns if str(c).lower() in {"split","set","partition"}), None)
assert idc and splitc, f"data_identification.csv must have id + split. Got: {list(ids.columns)}"
id_df = ids.rename(columns={idc:"id", splitc:"split"})[["id","split"]]
id_df["id"] = id_df["id"].astype(str)
id_df["split"] = id_df["split"].astype(str).str.lower()
print("Split counts:", id_df["split"].value_counts(dropna=False).to_dict())

#merging everything
df = posts_df.merge(id_df, on="id", how="inner").merge(emotion_df, on="id", how="left")

train_df = df[df["split"]=="train"].copy()
test_df  = df[df["split"]!="train"].copy()

#keeping a copy of the original label set (Kaggle)
original_labels = set(train_df["emotion"].dropna().unique().tolist())
print("Original Kaggle labels:", sorted(original_labels))

llm_df = None
llm_path = None
for cand in ["llm.csv", "llm.csv.csv"]:
    try:
        llm_path = find(cand)
        break
    except FileNotFoundError:
        llm_path = None

if llm_path is not None:
    print("Found LLM CSV:", llm_path)
    raw_llm = pd.read_csv(llm_path)

    #trying to infer id / text / emotion columns
    def _pick(colnames, keywords):
        for c in colnames:
            cl = str(c).lower()
            if any(k in cl for k in keywords):
                return c
        return None

    cols = list(raw_llm.columns)
    id_col_llm   = _pick(cols, ["id"])
    text_col_llm = _pick(cols, ["text","content","body"])
    emo_col_llm  = _pick(cols, ["emotion","label","target","class"])

    if id_col_llm is None or text_col_llm is None or emo_col_llm is None:
        print("Could not infer (id,text,emotion) from LLM CSV, skipping. Columns:", cols)
        llm_df = None
    else:
        llm_df = raw_llm.rename(
            columns={id_col_llm:"id", text_col_llm:"text", emo_col_llm:"emotion"}
        )[["id","text","emotion"]].dropna(subset=["text","emotion"]).copy()

        #making ids unique & tidy types
        llm_df["id"] = "llm_" + llm_df["id"].astype(str)
        llm_df["emotion"] = llm_df["emotion"].astype(str)

        #keeping only emotions that exist in the Kaggle training set
        before = len(llm_df)
        llm_df = llm_df[llm_df["emotion"].isin(original_labels)].copy()
        dropped = before - len(llm_df)
        if dropped > 0:
            print(f"Dropped {dropped} LLM rows with unseen emotions.")
        print("LLM rows kept:", len(llm_df))
else:
    print("No LLM CSV found → training will use only Kaggle data.")

#shared text cleaning
url_pat     = re.compile(r"http\S+|www\.\S+")
mention_pat = re.compile(r"@\w+")
name_pat    = re.compile(r"\[NAME\]", flags=re.IGNORECASE)
edit_pat    = re.compile(r"(edit:|tl;dr)", flags=re.IGNORECASE)
multi_space = re.compile(r"\s+")

def clean_text(s: str) -> str:
    s = str(s)

    #HTML unescape
    s = html.unescape(s)

    #replacing the URLs / mentions / [NAME] with special tokens
    s = url_pat.sub(" <URL> ", s)
    s = mention_pat.sub(" <USER> ", s)
    s = name_pat.sub(" <NAME> ", s)

    #removing leading quote lines
    lines = []
    for line in s.splitlines():
        if not line.lstrip().startswith(">"):
            lines.append(line)
    s = " ".join(lines)
    #truncate at edit markers
    m = edit_pat.search(s)
    if m:
        s = s[:m.start()]
    s = multi_space.sub(" ", s).strip()
    return s

#applying cleaning
train_df["text"] = train_df["text"].map(clean_text)
test_df["text"]  = test_df["text"].map(clean_text)

if llm_df is not None:
    llm_df["text"] = llm_df["text"].map(clean_text)
    #dropping any now-empty rows
    llm_df = llm_df[llm_df["text"].str.strip().ne("")].reset_index(drop=True)
    train_df = pd.concat([train_df, llm_df], ignore_index=True)
    print("Augmented train_df with LLM data →", train_df.shape)
else:
    print("No LLM augmentation used.")

MIN_WORDS = 3

def word_count(txt: str) -> int:
    return len(str(txt).split())

train_df["text_wc"] = train_df["text"].map(word_count)
before = len(train_df)
train_df = train_df[train_df["text_wc"] >= MIN_WORDS].copy()
after = len(train_df)
print(f"Dropped {before - after} very short training texts (<{MIN_WORDS} words).")
train_df = train_df.drop(columns=["text_wc"])

#removing duplicate texts (exact duplicates after cleaning)
train_df["text_norm"] = train_df["text"].str.lower().str.strip()
before = len(train_df)
train_df = train_df.drop_duplicates(subset=["text_norm"]).drop(columns=["text_norm"])
after = len(train_df)
print(f"Dropped {before - after} duplicate training texts after cleaning.")

#tidying columns
train_df = train_df[["id","text","emotion"]].dropna(subset=["text","emotion"]).reset_index(drop=True)
test_df  = test_df[["id","text"]].dropna(subset=["text"]).reset_index(drop=True)

print("Train shape:", train_df.shape, " Test:", test_df.shape)
print("Class counts:\n", train_df["emotion"].value_counts())
display(train_df.head(3))
display(test_df.head(3))
```

## 2. Feature Engineering Steps

```python
#imports i'll be using
import numpy as np
import torch
from torch.utils.data import Dataset
from transformers import AutoTokenizer, DataCollatorWithPadding

#just a placeholder
model_name = "cardiffnlp/twitter-roberta-base-emotion-latest"
max_len    = 256

#label maps
labels = sorted(train_df["emotion"].unique())
label2id = {lab: i for i, lab in enumerate(labels)}
id2label = {i: lab for lab, i in label2id.items()}

train_df["label"] = train_df["emotion"].map(label2id).astype(int)
y_np = train_df["label"].values  # make sure this matches the augmented train_df

print("Labels:", labels)
print(train_df["label"].value_counts())

#initial tokenizer...WILL BE OVERRIDDEN per-backbone in Model cell
tok = AutoTokenizer.from_pretrained(model_name, use_fast=True)
collator = DataCollatorWithPadding(tokenizer=tok)

class TextDataset(Dataset):
    def __init__(self, texts, labels=None):
        self.texts  = list(texts)
        self.labels = None if labels is None else np.array(labels, dtype=np.int64)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = tok(
            str(self.texts[idx]),
            truncation=True,
            max_length=max_len
        )
        if self.labels is None:
            return {**enc, "labels": None}
        return {**enc, "labels": int(self.labels[idx])}

#using augmented train_df + original test_df
ds_full = TextDataset(train_df["text"].tolist(), y_np)
ds_test = TextDataset(test_df["text"].tolist(), None)

print("Datasets ready:",
      f"train={len(ds_full)}",
      f"test={len(ds_test)}",
      f"num_labels={len(labels)}")
```

## 3. Model Implementation Steps

```python
import random, types, math, numpy as np, pandas as pd, torch
from typing import List, Optional, Tuple
from collections import Counter
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score
from torch.utils.data import Subset
import torch.nn as nn
from transformers import (
    AutoConfig, AutoModelForSequenceClassification, AutoTokenizer,
    Trainer, TrainingArguments, DataCollatorWithPadding, EarlyStoppingCallback
)
from dataclasses import dataclass

import os
os.environ["WANDB_DISABLED"] = "true"

#temp patch for environments where torch.version might be missing
if not hasattr(torch, "version"):
    torch.version = types.SimpleNamespace(cuda=None, hip=None)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

#seeding + global training setup
def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

#parameters [i borrowed a friend's lab env]
wd        = 0.05
bsz       = 16
grad_acc  = 2
nfolds    = 5
base_seed = 42

seed_everything(base_seed)

#three backbones
MODEL_BACKBONES = [
    "Amirhossein75/multi-label-emotion-classification-reddit-comments-roberta",
    "cardiffnlp/twitter-roberta-base-emotion-latest",
    "SamLowe/roberta-base-go_emotions",
]

HYPERPARAMETERS = {
    "Amirhossein75/multi-label-emotion-classification-reddit-comments-roberta": dict(lr=1.0e-5, epochs=4),
    "cardiffnlp/twitter-roberta-base-emotion-latest":                            dict(lr=1.3e-5, epochs=4),
    "SamLowe/roberta-base-go_emotions":                                          dict(lr=1.3e-5, epochs=4),
}

#label maps based on (Kaggle + LLM) train_df
labels    = sorted(train_df["emotion"].unique().tolist())
label2id  = {l: i for i, l in enumerate(labels)}
id2label  = {i: l for l, i in label2id.items()}
y_true_id = train_df["emotion"].map(label2id).values.astype(np.int64)
train_df["label"] = y_true_id  
#^^^ keeping numeric label on the dataframe too

#class weights
counts = Counter(y_true_id)
num_labels = len(labels)
class_weights = torch.ones(num_labels, dtype=torch.float32)
for i in range(num_labels):
    class_weights[i] = 1.0 / math.sqrt(counts.get(i, 1))
class_weights = class_weights / class_weights.mean()
print("Class weights:", class_weights.tolist())

#helper: weighted average ensemble
def weighted_average_ensemble(
    out_of_fold_list: List[np.ndarray],
    test_prediction_list: List[np.ndarray],
    weights: np.ndarray,
    ground_truth: Optional[np.ndarray] = None,
) -> Tuple[np.ndarray, np.ndarray]:
    oof_stack  = np.stack(out_of_fold_list, axis=0)
    test_stack = np.stack(test_prediction_list, axis=0)

    w = weights[:, None, None]

    x_out_of_fold = np.sum(oof_stack * w, axis=0)
    test_result   = np.sum(test_stack * w, axis=0)

    oof_pred = x_out_of_fold.argmax(axis=1)
    if ground_truth is not None:
        f1 = f1_score(ground_truth, oof_pred, average="macro")
        print(f"Weighted Ensemble OOF macro-F1: {f1:.4f}")

    return test_result, oof_pred

#CV folds (shared for all backbones) – use numeric labels for stratification
skf = StratifiedKFold(n_splits=nfolds, shuffle=True, random_state=base_seed)
fold_indices = list(skf.split(np.zeros(len(train_df)), y_true_id))

def subset(ds, idx):
    return Subset(ds, list(idx))

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=-1)
    return {"f1_macro": f1_score(labels, preds, average="macro")}

#custom Trainer with class-weighted loss
class WeightedTrainer(Trainer):
    def __init__(self, class_weights=None, *args, **kwargs):
        super().__init__(*args, **kwargs)
        if class_weights is not None:
            self.class_weights = class_weights.clone().detach()
        else:
            self.class_weights = None

    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        labels = inputs.get("labels")
        outputs = model(**{k: v for k, v in inputs.items() if k != "labels"})
        logits = outputs.logits

        if self.class_weights is not None:
            cw = self.class_weights.to(logits.device)
            loss_fct = nn.CrossEntropyLoss(weight=cw)
        else:
            loss_fct = nn.CrossEntropyLoss()

        loss = loss_fct(
            logits.view(-1, self.model.config.num_labels),
            labels.view(-1)
        )
        return (loss, outputs) if return_outputs else loss

#train each backbone and collect logits
all_oof_logits: List[np.ndarray]  = []
all_test_logits: List[np.ndarray] = []
backbone_names: List[str]         = []
backbone_oof_f1s: List[float]     = []

for backbone in MODEL_BACKBONES:
    print("\n" + "=" * 60)
    print(f"Training backbone: {backbone}")
    print("=" * 60)

    hp      = HYPERPARAMETERS.get(backbone, {})
    lr      = hp.get("lr", 1.5e-5)
    epochs  = hp.get("epochs", 4)

    #fresh tokenizer per backbone
    if "bertweet" in backbone.lower():
        tokenizer = AutoTokenizer.from_pretrained(backbone, use_fast=False)
    else:
        tokenizer = AutoTokenizer.from_pretrained(backbone, use_fast=True)

    #add special tokens like your friend
    special_tokens = ["<USER>", "<URL>", "<NAME>"]
    to_add = [t for t in special_tokens if t not in tokenizer.get_vocab()]
    num_added = 0
    if to_add:
        num_added = tokenizer.add_tokens(to_add)
        print(f"Added {num_added} special tokens:", to_add)

    #TextDataset uses global `tok`, so point it at this tokenizer
    global tok
    tok = tokenizer

    tr_texts  = train_df["text"].tolist()
    te_texts  = test_df["text"].tolist()
    tr_labels = y_true_id

    ds_full = TextDataset(tr_texts, tr_labels)
    ds_test = TextDataset(te_texts, None)

    @dataclass
    class SafeCollator(DataCollatorWithPadding):
        def __call__(self, features):
            cleaned = []
            for f in features:
                g = dict(f)
                for k in ("label", "labels"):
                    if k in g and g[k] is None:
                        g.pop(k)
                cleaned.append(g)
            return super().__call__(cleaned)

    collator = SafeCollator(tokenizer=tokenizer, pad_to_multiple_of=8)

    #storage for this backbone
    oof_logits = np.zeros((len(train_df), len(labels)), dtype=np.float32)
    test_logits_accum = None

    for fi, (tr_idx, va_idx) in enumerate(fold_indices):
        print(f"\n=== {backbone} — Fold {fi+1}/{nfolds} ===")
        tr_ds = subset(ds_full, tr_idx)
        va_ds = subset(ds_full, va_idx)

        cfg = AutoConfig.from_pretrained(
            backbone,
            num_labels=len(labels),
            id2label=id2label,
            label2id=label2id,
            attn_implementation="eager",
            problem_type="single_label_classification"
        )
        model = AutoModelForSequenceClassification.from_pretrained(
            backbone,
            config=cfg,
            ignore_mismatched_sizes=True,
        ).to(device)

        #resize embeddings if we added special tokens
        if num_added > 0:
            model.resize_token_embeddings(len(tokenizer))

        args = TrainingArguments(
            output_dir=f"./out_{backbone.replace('/', '_')}_fold{fi}",
            learning_rate=lr,
            per_device_train_batch_size=bsz,
            per_device_eval_batch_size=bsz * 4,
            gradient_accumulation_steps=grad_acc,
            num_train_epochs=epochs,
            weight_decay=wd,
            warmup_ratio=0.10,
            lr_scheduler_type="cosine",
            max_grad_norm=0.5,
            logging_steps=50,
            eval_strategy="epoch",
            save_strategy="epoch",
            load_best_model_at_end=True,
            metric_for_best_model="f1_macro",
            greater_is_better=True,
            fp16=(device == "cuda"),
            gradient_checkpointing=False,
            seed=base_seed + fi,
            report_to="none",
            label_smoothing_factor=0.05,
            dataloader_num_workers=2,
            dataloader_pin_memory=(device=="cuda"),
            save_total_limit=1,
        )

        trainer = WeightedTrainer(
            class_weights=class_weights,
            model=model,
            args=args,
            train_dataset=tr_ds,
            eval_dataset=va_ds,
            tokenizer=tokenizer,
            data_collator=collator,
            compute_metrics=compute_metrics,
            callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],
        )

        trainer.train()

        #validation predictions → probabilities
        va_out    = trainer.predict(va_ds)
        va_logits = va_out.predictions
        va_probs  = torch.softmax(torch.tensor(va_logits), dim=-1).numpy()
        oof_logits[va_idx] = va_probs

        # test predictions → probabilities, accumulate across folds
        te_out    = trainer.predict(ds_test)
        te_logits = te_out.predictions
        te_probs  = torch.softmax(torch.tensor(te_logits), dim=-1).numpy()
        test_logits_accum = te_probs if test_logits_accum is None else (test_logits_accum + te_probs)

    #average test probs over folds for this backbone
    test_probs_mean = test_logits_accum / nfolds

    #store for later ensembling
    all_oof_logits.append(oof_logits)
    all_test_logits.append(test_probs_mean)

    #quick sanity check per backbone
    oof_pred_ids = oof_logits.argmax(axis=-1)
    oof_f1 = f1_score(y_true_id, oof_pred_ids, average="macro")
    print(f"Backbone {backbone} OOF macro-F1: {oof_f1:.4f}")

    backbone_names.append(backbone)
    backbone_oof_f1s.append(oof_f1)
```

## 3. Ensembler

```python
#ensembles the csv
from typing import Dict, List, Tuple
import os

import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


def _compute_oof_f1_scores(
    oof_logits_list: List[np.ndarray],
    y_true: np.ndarray,
) -> np.ndarray:
    """
    Helper: compute macro-F1 scores for each model's OOF predictions.
    Returns an array of shape [num_models].
    """
    scores = []
    for i, oof_logits in enumerate(oof_logits_list):
        pred = oof_logits.argmax(axis=1)
        f1 = f1_score(y_true, pred, average="macro")
        print(f"Model {i} → OOF macro-F1 = {f1:.4f}")
        scores.append(f1)
    return np.asarray(scores, dtype=np.float32)


def _normalize_weights(raw_scores: np.ndarray, power: float = 1.0) -> np.ndarray:
    """
    Turn raw scores into normalized, positive weights.

    power > 1.0 sharpens differences between strong/weak models.
    """
    scores = np.maximum(raw_scores, 0.0)
    if power != 1.0:
        scores = scores**power
    total = scores.sum()
    if total == 0.0:
        return np.ones_like(scores) / len(scores)
    return scores / total

#f1 weight balancing
def f1_weighted_blend(
    oof_logits_list: List[np.ndarray],
    test_logits_list: List[np.ndarray],
    y_true: np.ndarray,
    weight_power: float = 1.0,
) -> Tuple[np.ndarray, np.ndarray, float]:
    """
    Blend model logits by weighting each model according to its own
    macro-F1 score on OOF predictions.

    Parameters
    ----------
    oof_logits_list : list of (N_train, num_classes)
    test_logits_list : list of (N_test, num_classes)
    y_true : (N_train,)
        True labels for OOF predictions.
    weight_power : float
        Exponent applied to F1 scores before normalization. Larger values
        emphasize high-performing models more strongly.
    """
    #per-model oof f1
    per_model_f1 = _compute_oof_f1_scores(oof_logits_list, y_true)
    weights = _normalize_weights(per_model_f1, power=weight_power)
    print("Ensemble weights (from OOF F1):", weights.tolist())

    #stack logits into arrays of shape
    oof_stack = np.stack(oof_logits_list, axis=0)
    test_stack = np.stack(test_logits_list, axis=0)

    #weighted sum along model axis using tensordot
    blended_oof = np.tensordot(weights, oof_stack, axes=(0, 0))
    blended_test = np.tensordot(weights, test_stack, axes=(0, 0))

    #evaluate blended OOF
    oof_pred = blended_oof.argmax(axis=1)
    f1 = f1_score(y_true, oof_pred, average="macro")
    print(f"[F1-weighted blend] macro-F1 = {f1:.4f}")

    return blended_test, oof_pred, f1

#logistic Regression stacking
def logistic_stacking(
    oof_logits_list: List[np.ndarray],
    test_logits_list: List[np.ndarray],
    y_true: np.ndarray,
    seed: int = 42,
    C: float = 0.1,
    max_iter: int = 1000,
) -> Tuple[np.ndarray, np.ndarray, float]:
    """
    Stacking ensemble: train a multinomial Logistic Regression meta-model
    on concatenated logits from all base models.
    """
    #concatenate logits along the feature dimension
    meta_train = np.hstack(oof_logits_list)
    meta_test = np.hstack(test_logits_list)

    #meta-learner
    meta_clf = LogisticRegression(
        max_iter=max_iter,
        C=C,
        random_state=seed,
        n_jobs=None,
    )
    meta_clf.fit(meta_train, y_true)

    #oof performance of the stacker
    oof_pred = meta_clf.predict(meta_train)
    f1 = f1_score(y_true, oof_pred, average="macro")
    print(f"[Logistic stacking] macro-F1 = {f1:.4f}")

    #test probabilities
    test_proba = meta_clf.predict_proba(meta_test)
    return test_proba, oof_pred, f1

#running both ensembles + save CSVs and confusion matrices
def run_weighted_and_stacked(
    oof_logits_list: List[np.ndarray],
    test_logits_list: List[np.ndarray],
    y_true: np.ndarray,
    test_df: pd.DataFrame,
    id_to_label: Dict[int, str],
    out_dir: str = "./submission",
) -> None:
    """
    Compare F1-weighted blending vs stacking and export:
      - submission_<method>.csv
      - confusion_matrix_<method>.png
    """
    os.makedirs(out_dir, exist_ok=True)
    cm_dir = os.path.join(out_dir, "confusion_matrix")
    os.makedirs(cm_dir, exist_ok=True)

    results: Dict[str, Tuple[float, np.ndarray, np.ndarray]] = {}

    #f1-weighted blending
    test_weighted, oof_weighted, f1_weighted = f1_weighted_blend(
        oof_logits_list=oof_logits_list,
        test_logits_list=test_logits_list,
        y_true=y_true,
        weight_power=1.0,
    )
    results["weighted"] = (f1_weighted, oof_weighted, test_weighted)

    #logistic stacking
    test_stacked, oof_stacked, f1_stacked = logistic_stacking(
        oof_logits_list=oof_logits_list,
        test_logits_list=test_logits_list,
        y_true=y_true,
        seed=42,
        C=0.1,
        max_iter=1000,
    )
    results["stacked"] = (f1_stacked, oof_stacked, test_stacked)

    # pick best method by OOF macro-F1
    best_name = max(results, key=lambda k: results[k][0])
    best_f1 = results[best_name][0]
    print(f"\n>>> Best ensemble (OOF macro-F1): {best_name} = {best_f1:.4f}\n")

    #exporting submissions + confusion matrices for each method [stacked and weight...did have avg before but it was ineff]
    label_order = [id_to_label[idx] for idx in sorted(id_to_label.keys())]

    for name, (f1_val, oof_pred, test_proba) in results.items():
        #safety check
        if test_proba.shape[0] != len(test_df):
            raise ValueError(
                f"[{name}] prediction size ({test_proba.shape[0]}) "
                f"does not match test_df length ({len(test_df)})"
            )

        #test-set hard labels
        final_pred = test_proba.argmax(axis=1)
        pred_labels = [id_to_label[idx] for idx in final_pred]

        submission = pd.DataFrame(
            {
                "id": test_df["id"].values,
                "emotion": pred_labels,
            }
        )
        sub_path = os.path.join(out_dir, f"submission_{name}.csv")
        submission.to_csv(sub_path, index=False)
        print(f"[{name}] Saved submission → {sub_path} (OOF macro-F1 = {f1_val:.4f})")

        #confusion matrix on OOF predictions
        cm = confusion_matrix(y_true, oof_pred)
        disp = ConfusionMatrixDisplay(
            confusion_matrix=cm,
            display_labels=label_order,
        )
        disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
        plt.title(f"Confusion Matrix – {name}")
        plt.tight_layout()

        cm_path = os.path.join(cm_dir, f"confusion_matrix_{name}.png")
        plt.savefig(cm_path)
        plt.close()
        print(f"[{name}] Saved confusion matrix → {cm_path}")


#calling to generate csv and matric
run_weighted_and_stacked(
    oof_logits_list=all_oof_logits,
    test_logits_list=all_test_logits,
    y_true=y_true_id,
    test_df=test_df,
    id_to_label=id2label,
    out_dir="./submission",
)
```
